# Extreme value theory

```python
a = np.linspace(-4., 4., num=1000)
```

```python
bg = np.exp(-np.exp(-a)) # Gumbel aka logistic
```

```python
alpha = 1.5
```

```python
bf = [] # Fréchet

for i in xrange(len(a)):
    if a[i] <= 0:
        bf.append(0.)
    else:
        bf.append(np.exp(-(a[i] ** (-alpha))))
```

```python
bw = [] # Weibull

for i in xrange(len(a)):
    if a[i] <= 0:
        bw.append(np.exp(-((-a[i]) ** alpha))) #np.exp(-(-a[i] ** alpha)))
    else:
        bw.append(1.)
```

```python
bw
```

```python
plt.plot(a,bw)
plt.plot(a,bg)
plt.plot(a,bf)
```

```python
#generalized extreme value (GEV)

c = np.linspace(-10., 10., num=100)

b = [] 

for j in xrange(len(c)):
    d = []
    for i in xrange(len(a)):
        if c[j] != 0:
            d.append(np.exp(-((1. + c[j] * a[i]) ** (-1./c[j])))) 
        else:
            d.append(np.exp(-np.exp(-a[i])))
            
    b.append(d)
```

```python
plt.plot(a,b[0])
plt.plot(a,b[9])
plt.plot(a,b[19])
plt.plot(a,b[29])
plt.plot(a,b[39])
plt.plot(a,b[49])
plt.plot(a,b[59])
plt.plot(a,b[69])
plt.plot(a,b[79])
plt.plot(a,b[89])
plt.plot(a,b[99])
```

```python
#Peak of threshold (POT)

sigma = 1.

c = np.linspace(-10., 10., num=100)

b = [] 

for j in xrange(len(c)):
    d = []
    for i in xrange(len(a)):
        if c[j] != 0:
            d.append(1. - ((1. + c[j]/sigma * a[i]) ** (-1./c[j]))) 
        else:
            d.append(1. - np.exp(-a[i]/sigma))
            
    b.append(d)
```

```python
plt.plot(a,b[0])
plt.plot(a,b[9])
plt.plot(a,b[19])
plt.plot(a,b[29])
plt.plot(a,b[39])
plt.plot(a,b[49])
plt.plot(a,b[59])
plt.plot(a,b[69])
plt.plot(a,b[79])
plt.plot(a,b[89])
plt.plot(a,b[99])
```
