```python
# block maxima

m = len(prices['SNPLogRtn'])/252

a = [] #min

b = [] #max

for i in xrange(252):
    tmp = np.array(prices['SNPLogRtn'][(i * m):(((i+1) *m)-1)])
    a.append(np.min(tmp))
    b.append(np.max(tmp))
```

```python
# Hill plot - left tail
```

```python
x = np.array(-prices['SNPLogRtn'][prices['SNPLogRtn']<0.])

x = np.sort(x)[::-1]

x
```

```python
bl = [] # Hill estimate 
a = []

n = len(x)

for i in xrange(n):    
    a.append(x[i])
    c = []
    for j in xrange(len(a)):
        c.append(np.log(a[j]))
    bl.append([i, 1./(np.mean(c) - np.log(x[i]))]) 
```

```python
bl1 = [bl[i][1] for i in xrange(len(bl))]
k = [bl[i][0] for i in xrange(len(bl))]
```

```python
plt.plot(k[:2000], bl1[:2000],linewidth=.5)
plt.xticks(np.linspace(0, 2000, 21, endpoint=True))
plt.yticks(np.linspace(0, 7., 21, endpoint=True))
```

```python
x[200]
```

```python
plt.plot(x, bl1,linewidth=.5)

plt.xticks(np.linspace(0, 10, 26, endpoint=True))
#plt.yticks(np.linspace(0, 7, 21, endpoint=True))
```

```python
# Hill plot - right tail
y = np.array(-prices['SNPLogRtn'][prices['SNPLogRtn']>0.])

y = np.sort(x)[::-1]



br = [] # Hill estimate 
a = []

for i in xrange(len(y)-1):
    a.append(y[i])
    c = []
    for j in xrange(len(a)):
        c.append(np.log(a[j]))
    br.append([i, 1./(np.mean(c) - np.log(y[i] ))])
    
br1 = [br[i][1] for i in xrange(len(br))]
k = [br[i][0] for i in xrange(len(br))]    
    
plt.plot(k[:2000], br1[:2000],linewidth=.5)
plt.xticks(np.linspace(0, 2000, 21, endpoint=True))
plt.yticks(np.linspace(0, 7., 21, endpoint=True))
```

```python
# alternative Hill plot - left tail

n = len(bl)
m = 1000

d = np.linspace(0., 1., num=m)

el = []

for i in xrange(m):
    
    f = 0
    for j in xrange(n):
        if n**d[i] >= j: 
            f = j
            
    el.append(bl1[f])

```

```python
plt.plot(d, el,linewidth=.5)
plt.xticks(np.linspace(0, 1., 21, endpoint=True))
plt.yticks(np.linspace(0, 7., 21, endpoint=True))
```

```python
# alternative Hill plot - right tail

n = len(br)
m = 1000

d = np.linspace(0., 1., num=m)

er = []

for i in xrange(m):
    
    f = 0
    for j in xrange(n):
        if n**d[i] >= j: 
            f = j
            
    er.append(br1[f])
    
plt.plot(d, er,linewidth=.5)
plt.xticks(np.linspace(0, 1., 21, endpoint=True))
plt.yticks(np.linspace(0, 7., 21, endpoint=True))
```

```python
# percentage Hill - left tail 

n = len(bl)
epsilon = 1.e-2

h = []


for i in xrange(n):

    g = 0.
    for j in xrange(n):
        if np.absolute(bl1[j] - bl1[i]) <= epsilon:
            g += 1
    
    h.append(g/n)
    
    
```

```python
plt.plot(bl1,h, 'bo')
```

```python
# percentage Hill - right tail 

n = len(br)
epsilon = 1.e-2

h = []


for i in xrange(n):

    g = 0.
    for j in xrange(n):
        if np.absolute(br1[j] - br1[i]) <= epsilon:
            g += 1
    
    h.append(g/n)
    
```
